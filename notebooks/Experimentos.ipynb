```python
embeddings = OllamaEmbeddings(model = "nomic-embed-text")
ejemplo = ["Las politicas del producto financiero CPS son que solo se le puede otorgar a personas con una linea CPR activa.",
            "Las politicas del producto CPR es que se debe de cumplir que el salario de la persona debera de ser al menos del 40% del valor de la linea CPR."]

db = Chroma.from_texts(ejemplo, embeddings, persist_directory = "./db")

def internal_documentation_info(query : str):
    resultados = db.similarity_search(query, k = 1)
    return "\n".join([r.page_content for r in resultados])


from langchain.tools import tool

@tool
def internal_documentation_info_tool(query : str):
    """
    Busca información técnica detallada en la documentación interna 
    de la empresa sobre un tema específico.
    """
    return internal_documentation_info(query)
```

```python
from langchain.agents import create_agent
from langchain.tools import tool
from langchain_anthropic import ChatAnthropic
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import Chroma
```

```python

claude = ChatAnthropic(
    model="claude-opus-4-7",
    effort="xhigh",
    max_tokens=90,
    api_key= 'api_key'
)

agente = create_agent(
    tools = [internal_documentation_info_tool],
    model = claude,
    system_prompt = ("Eres un asistente para responder preguntas especifcias. ",
                     "Prioriza la informacion verificable. ",
                     "Utiliza herramientas cuando se te pida informacion especifica de la empresa. ",
                     "No inventes informacion. ")
)

prompt_agente = "Que necesito para tener una linea CPR?"
    
resultado = agente.invoke(    
    {"messages" : [{"role": "user", "content": prompt_agente}]}
    )
```
