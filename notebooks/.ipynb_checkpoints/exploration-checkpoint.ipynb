```python
categorical_columns = []
numerical_columns = []
dates_columns = []

for col in df.columns:
    if df[col].dtype == "o" or df[col].dtype == "str":
        print(f"for column {col}, unique values are: \n {df[col].unique()} \n")
        categorical_columns.append(col)
        continue
    elif df[col].dtype == "<M8[us]":
        dates_columns.append(col)
    else:
        numerical_columns.append(col)

print("# numerical columns: ", len(numerical_columns))
```

```python
zero_columns = ['Customers', 'Open', 'Promo', 'SchoolHoliday']
categorical_columns = categorical_columns + ['DayOfWeek']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import RFECV
from sklearn.model_selection import cross_val_score, train_test_split, TimeSeriesSplit
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.ensemble import GradientBoostingRegressor

class DayImputer(BaseEstimator, TransformerMixin):

    def __init__(self, date_col, day_col):
        self.date_col = date_col
        self.day_col =  day_col

    def fit(self, X, y = None):
        return self

    def transform(self, X):
        X = X.copy()
        dias = pd.to_datetime(X[self.date_col]).dt.weekday + 1
        X.loc[X[self.day_col].isna(), self.day_col] =  dias

        X = X.drop(columns = [self.date_col])
        return X

def add_time_features(df : pd.DataFrame, cols_dates : list) -> pd.DataFrame:
    for col in cols_dates:
        df[f"day_{col}"] = df[col].dt.weekday
        df[f"week_{col}"] = df[col].dt.isocalendar().week.astype(int)
        df[f"month_{col}"] = df[col].dt.month
        df[f"is_weekend_{col}"] = (df[f"day_{col}"] >= 5).astype(int)
        df[f"year_{col}"] = df[col].dt.year
    
    return df


def add_lag_features(df : pd.DataFrame, lags = [1, 7, 14]) -> pd.DataFrame:
    df = df.copy()
    for lag in lags:
        df[f"Sales_lag_{lag}"] = (df.groupby("Store")["Sales"].shift(lag))
    return df
    

def add_rolling_features(df : pd.DataFrame, windows = [7, 14]) -> pd.DataFrame:
    df = df.copy()
    for w in windows:
        df[f"Sales_roll_mean_{w}"] = (df.groupby("Store")["Sales"].shift(1).rolling(w).mean())
    return df
    

num_pipeline = Pipeline([("imputer", SimpleImputer(strategy = "constant", fill_value = 0)), 
                           ("scaler", StandardScaler())])

categorical_pipeline = Pipeline([("imputer", SimpleImputer(strategy = "constant", fill_value = "0")), 
                                 ("onehot", OneHotEncoder(handle_unknown = "ignore"))])

preprocessor = ColumnTransformer([
    ("num", num_pipeline, zero_columns),
    ("cat", categorical_pipeline, categorical_columns)
], remainder = "passthrough")

base_model = GradientBoostingRegressor()

outer_tscv = TimeSeriesSplit(n_splits = 5)
inner_tscv = TimeSeriesSplit(n_splits = 3)

rfecv = RFECV(estimator = base_model, step = 1, cv = inner_tscv, scoring = "neg_mean_absolute_error", n_jobs = -1, verbose = 0)

full_pipeline = Pipeline([
    ("day_imputer", DayImputer('Date', 'DayOfWeek')),
    ("preprocessor", preprocessor),
    ("feature_selection", rfecv)
])
```

```python
df = add_time_features(df, ["Date"])

df = df.sort_values("Date").reset_index(drop = True)
tarjet_col = "Sales"

results = []
for fold, (train_idx, test_idx) in enumerate(outer_tscv.split(df)):

    train_df = df.iloc[train_idx].copy()
    test_df = df.iloc[test_idx].copy()

    train_df = add_lag_features(train_df,lags = [1, 7, 14])
    train_df = add_rolling_features(train_df, windows = [1, 7, 14])

    max_lag = 14
    combined = pd.concat([train_df.tail(max_lag), test_df], axis = 0)
    combined = add_lag_features(combined, lags = [1, 7, 14])

    test_df = combined.iloc[max_lag:].copy()

    train_df = train_df.dropna().reset_index(drop = True)
    test_df = test_df.dropna().reset_index(drop = True)

    X_train = train_df.drop(columns = [tarjet_col])
    y_train = train_df[tarjet_col]
    
    X_test = test_df.drop(columns = [tarjet_col])
    y_test = test_df[tarjet_col]

    full_pipeline.fit(X_train, y_train)

    results.append(full_pipeline.named_steps["feature_selection"].get_feature_names_out())
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import cross_val_score, train_test_split, TimeSeriesSplit, GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor


pass_columns = ['Customers', 'Promo']

INTERMEDIATE = Path("../data/intermediate/intermediate_data.csv")

df = pd.read_csv(INTERMEDIATE, low_memory = False, parse_dates = ["Date"])


def add_lag_features(df : pd.DataFrame, lags = [1, 7, 14]) -> pd.DataFrame:
    df = df.copy()
    for lag in lags:
        df[f"Sales_lag_{lag}"] = (df.groupby("Store")["Sales"].shift(lag))
    return df
    

def add_rolling_features(df : pd.DataFrame, windows = [1]) -> pd.DataFrame:
    df = df.copy()
    for w in windows:
        df[f"Sales_roll_mean_{w}"] = (df.groupby("Store")["Sales"].shift(1).rolling(w).mean())
    return df
    

num_pipeline = Pipeline([("imputer", SimpleImputer(strategy = "constant", fill_value = 0)), 
                         ("scaler", StandardScaler())])

preprocessor = ColumnTransformer([
    ("num", num_pipeline, pass_columns)
    ])

base_model = GradientBoostingRegressor()

outer_tscv = TimeSeriesSplit(n_splits = 5)
inner_tscv = TimeSeriesSplit(n_splits = 3)

param_grid = {
    "n_estimators" : [100, 200, 300, 500, 1000],
    "learning_rate" : [0.01, 0.05, 0.1],
    "max_depth" : [3, 4, 5, 7, 10]
}

grid = GridSearchCV(estimator = base_model, param_grid = param_grid, cv = inner_tscv, scoring = "neg_mean_absolute_error", n_jobs = -1, verbose = 0)

grid_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("grid_selection", grid)
])
```

```python
df = df.sort_values("Date").reset_index(drop = True)
tarjet_col = "Sales"

results_grid_params = []
results_grid_score = []
for fold, (train_idx, test_idx) in enumerate(outer_tscv.split(df)):

    train_df = df.iloc[train_idx].copy()
    test_df = df.iloc[test_idx].copy()

    train_df = add_lag_features(train_df,lags = [1, 7, 14])
    train_df = add_rolling_features(train_df, windows = [1])

    max_lag = 14
    combined = pd.concat([train_df.tail(max_lag), test_df], axis = 0)
    combined = add_lag_features(combined, lags = [1, 7, 14])

    test_df = combined.iloc[max_lag:].copy()

    train_df = train_df.dropna().reset_index(drop = True)
    test_df = test_df.dropna().reset_index(drop = True)

    X_train = train_df.drop(columns = [tarjet_col])
    y_train = train_df[tarjet_col]
    
    X_test = test_df.drop(columns = [tarjet_col])
    y_test = test_df[tarjet_col]

    grid_pipeline.fit(X_train, y_train)

    results_grid_params.append(grid_pipeline.named_steps["grid_selection"].best_params_)
    results_grid_score.append(grid_pipeline.named_steps["grid_selection"].best_score_)
```
